## Import Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
```

### Functions

```python
def scale_data(X):
    scaler = preprocessing.MinMaxScaler()
    scaler.fit(X)

    return pd.DataFrame(scaler.transform(X), columns=X.columns)
```

## Data Preperation

### Import Data

```python
path = "Data/diabetes.csv"
data = pd.read_csv(path)
data.head()
```

### Split Data Into Train, Validation and Test Sets

```python
# The target size for each set
train_size = 0.7
val_size = 0.15
test_size = 0.15

# Split the data into features (X) and the target (y)
X, y = data.drop(columns=["Outcome"]), data[["Outcome"]]

# Split into train_validation set and test set
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=test_size, random_state=1)

# Split the train_validation set into train sets and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=(val_size/(val_size + train_size)), random_state=1)
```

```python
print(f"Size of Train set {len(X_train)}, as a percentage {np.round(len(X_train)/len(X),3)}%")
print(f"Size of Validation set {len(X_val)}, as a percentage {np.round(len(X_val)/len(X),3)}%")
print(f"Size of Test set {len(X_test)}, as a percentage {np.round(len(X_test)/len(X),3)}%")
```

### Preprocess Data

```python
X_train.info()
```

```python
X_train.describe()
```

BMI is calculated as;  
**BMI** = weight(KG) / ((height(Meters)^2))  
So the only way someones BMI is 0, is if they weigh 0KG.   
For this reason I will assume that this an error in the data and remove any rows with the value as 0.

**Blood pressure** ([BP](https://en.wikipedia.org/wiki/Blood_pressure)) is the pressure of circulating blood against the walls of blood vessels.  
From this defination you would assume that someone has to have a BP > 0 to be alive.

**Skin Thickness** can not be less than or equal to 0mm thick. For [triceps skin-fold](https://medical-dictionary.thefreedictionary.com/triceps+skin-fold+thickness#:~:text=A%20value%20used%20to%20estimate,%3B%20in%20females%2C%2023%20mm.) thickness the normal thickness in males is 12 mm; in females, 23 mm.


```python
X_train_copy = X_train.copy(deep = True)
X_val_copy = X_val.copy(deep = True)
X_test_copy = X_test.copy(deep = True)

X_train_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = X_train_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
X_val_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = X_val_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
X_test_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = X_test_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)


## showing the count of Nans
print("The null count for the train set", end="\n\n")
print(X_train_copy.isnull().sum(), end="\n\n")
print("The null count for the val set", end="\n\n")
print(X_val_copy.isnull().sum(), end="\n\n")
print("The null count for the test set", end="\n\n")
print(X_test_copy.isnull().sum(), end="\n\n")
```

The two methods that I will use to fill in the missing values are:

1. Fill the null values with either the mean or mode for the value depending on the distribution.
2. Estimate the value using the K-Means algorithm.

```python
sns.set_theme(style="ticks")
sns.pairplot(X_train_copy.join(y_train), hue="Outcome");
```

#### Filling missing values using averages (Mean / Mode)

```python
index_map = {"mean":"Mean",
          "50%":"Median",
          0:"Mode",}

X_train_dist = X_train_copy.describe().loc[["mean","50%"]].append(X_train_copy.mode().iloc[0]).rename(index=index_map)

X_train_dist[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]
```

```python
X_train_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].hist(figsize = (20,20));
```

```python
X_train_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].skew(axis = 0, skipna = True)
```

```python
# I will use the mean for the features that follow similiar distributions to the normal distribution.
avg_Glucose =  X_train_copy['Glucose'].mean() 
avg_BloodPressure =  X_train_copy['BloodPressure'].mean()
avg_SkinThickness =  X_train_copy['SkinThickness'].mean()
avg_BMI =  X_train_copy['BMI'].mean()

avg_Insulin =  X_train_copy['Insulin'].median()
```

```python
# Fill the NaN values using the average found using the train set
X_train_fillna_avg, X_val_fillna_avg, X_test_fillna_avg = X_train_copy.copy(deep=True), X_val_copy.copy(deep=True), X_test_copy.copy(deep=True)

X_train_fillna_avg['Glucose'] = X_train_copy['Glucose'].fillna(avg_Glucose)
X_train_fillna_avg['BloodPressure'] = X_train_copy['BloodPressure'].fillna(avg_BloodPressure)
X_train_fillna_avg['SkinThickness'] = X_train_copy['SkinThickness'].fillna(avg_SkinThickness)
X_train_fillna_avg['BMI'] = X_train_copy['BMI'].fillna(avg_BMI)
X_train_fillna_avg['Insulin'] = X_train_copy['Insulin'].fillna(avg_Insulin)

X_val_fillna_avg['Glucose'] = X_val_copy['Glucose'].fillna(avg_Glucose)
X_val_fillna_avg['BloodPressure'] = X_val_copy['BloodPressure'].fillna(avg_BloodPressure)
X_val_fillna_avg['SkinThickness'] = X_val_copy['SkinThickness'].fillna(avg_SkinThickness)
X_val_fillna_avg['BMI'] = X_val_copy['BMI'].fillna(avg_BMI)
X_val_fillna_avg['Insulin'] = X_val_copy['Insulin'].fillna(avg_Insulin)

X_test_fillna_avg['Glucose'] = X_test_copy['Glucose'].fillna(avg_Glucose)
X_test_fillna_avg['BloodPressure'] = X_test_copy['BloodPressure'].fillna(avg_BloodPressure)
X_test_fillna_avg['SkinThickness'] = X_test_copy['SkinThickness'].fillna(avg_SkinThickness)
X_test_fillna_avg['BMI'] = X_test_copy['BMI'].fillna(avg_BMI)
X_test_fillna_avg['Insulin'] = X_test_copy['Insulin'].fillna(avg_Insulin)
```

```python
# Scale using MinMaxScaler, I scaled separately to avoid each set effecting eachother.
X_train_fillna_avg_scaled = scale_data(X_train_fillna_avg)
X_val_fillna_avg_scaled = scale_data(X_val_fillna_avg)
X_test_fillna_avg_scaled = scale_data(X_test_fillna_avg)
```

## Random Forest classifier

#### Filled NaN values using averages

```python
clf = RandomForestClassifier(max_depth=2, random_state=0,)
clf.fit(X_train_fillna_avg_scaled, y_train)

yhat_val = clf.predict(X_val_fillna_avg_scaled)

print(confusion_matrix(yhat_val, y_val))
print(accuracy_score(yhat_val, y_val))
print(f1_score(yhat_val, y_val))

yhat_test = clf.predict(X_test_fillna_avg_scaled)

print(confusion_matrix(yhat_test, y_test))
print(accuracy_score(yhat_test, y_test))
print(f1_score(yhat_test, y_test))
```
